```python
#var gislog()
from bs4 import BeautifulSoup
import re
import requests 
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
from datetime import datetime
from datetime import date,timedelta

url = 'https://bitinfocharts.com/comparison/bitcoin-transactions.html' 
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html5lib')   
```

```python
data_string = soup.findAll('script')[5].get_text()
```

```python
output = data_string.split('{}')
```

```python
tmp = output[0].split('"container"),')[1]
```

```python
block = tmp.split(', {labels:')[0]
```

```python
strs = block.replace('[','').split('],')
lists = [list(map(str, s.replace(']','').split(','))) for s in strs]
```

```python
transactions = []
def to_datetime(date_str):
    ds = date_str.replace('(',')').replace('\"','').split(')')[1]
    date = datetime.strptime(ds,'%Y/%m/%d')
    return date
dates = []
for i in range(0,len(lists)):
    tmp = lists[i][1]
    if tmp != 'null':
        transactions.append(int(tmp))
        dates.append(to_datetime(lists[i][0]))

tweets_dict = {}
tweets_dict['transactions'] = transactions
tweets_dict['dates'] = dates
```

```python
transactions_df = pd.DataFrame(tweets_dict)
```
